```python
import matplotlib.pyplot as plt

# Degree (Ground Truth)
degrees = [-60, -45, -30, 0, 30, 45, 60]

# PDOA values
pdoa_gt = [-2.7207, -2.2214, -1.5708, 0, 1.5708, 2.2214, 2.7207]
mean_pdoa = [-2.09, -1.77, -1.33, 0, 1.34, 2.43, 2.60]

# AOA values
aoa_gt = degrees  # Ground Truth AOA is the same as degree
mean_aoa = [-47.95, -34.24, -25.04, 0.04, 25.21, 50.63, 56.18]

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot for PDOA
ax[0].plot(degrees, pdoa_gt, 'o-', label='Ground Truth PDOA')
ax[0].plot(degrees, mean_pdoa, 's--', label='Mean PDOA')
ax[0].set_title('PDOA Comparison')
ax[0].set_xlabel('Degree')
ax[0].set_ylabel('PDOA')
ax[0].legend()
ax[0].grid(True)

# Plot for AOA
ax[1].plot(degrees, aoa_gt, 'o-', label='Ground Truth AOA')
ax[1].plot(degrees, mean_aoa, 's--', label='Mean AOA')
ax[1].set_title('AOA Comparison')
ax[1].set_xlabel('Degree')
ax[1].set_ylabel('AOA (Degrees)')
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
